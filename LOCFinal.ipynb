```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

news_1 = []

for i in range(1, 6):
    # URL of the website you want to scrape
    url = f'https://www.livemint.com/mutual-fund/page-{i}'

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the div section containing the articles by class
        articles_div = soup.find('section', class_='mainSec')

        # Check if the articles_div is not None
        if articles_div:
            # Find all <a> tags within the div
            a_tags = articles_div.find_all('a')
            # Loop through each <a> tag
            for a_tag in a_tags:
                # Extract the URL from the href attribute
                href = a_tag.get('href')
                # Check if href is not None and not starts with 'javascript:'
                if href and not href.startswith('javascript:'):
                    # Join the extracted URL with the base URL to handle relative URLs
                    full_url = urljoin(url, href)
                    # Send a GET request to the full URL
                    linked_response = requests.get(full_url)
                    # Check if the request was successful
                    if linked_response.status_code == 200:
                        # Parse the HTML content of the linked page
                        linked_soup = BeautifulSoup(linked_response.content, 'html.parser')
                        # Find the <p> tag containing the desired content
                        p_tag = linked_soup.find_all('p')
                        p_text = ' '.join(p_tag.text.strip() for p_tag in p_tag)
                        # If the <p> tag is found, extract its text content
                        if p_tag:
                            # Append the <a> tag's text content and the <p> tag's text content to news_1
                            news_1.append({'a_tag': a_tag.text.strip(), 'p_tag': p_text})
                    else:
                        print(f"Failed to fetch data from URL: {full_url}")
    else:
        print(f"Failed to fetch data from URL: {url}")

news_1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
stopWords = stopwords.words('english')
cv = CountVectorizer(stop_words=stopWords, ngram_range=(2, 3), max_features=10)
x = cv.fit_transform(cleaned_t)
print(cv.vocabulary_)
```

```python
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(cv.vocabulary_)

num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(tfidf_matrix)
clusters = kmeans.labels_


G = nx.Graph()
for i, text in enumerate(cv.vocabulary_):
    cluster = clusters[i]
    G.add_node(text, cluster=cluster, text=text)

threshold = -1

for i, a in enumerate(cv.vocabulary_):
    for j, b in enumerate(cv.vocabulary_):
        similarity = cosine_similarity(tfidf_matrix[i], tfidf_matrix[j])[0][0]
        if similarity > threshold:
          print(similarity)
          G.add_edge(a, b, weight=similarity)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Plot the network graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, k=0.1)  # Adjust the spring_layout parameters as needed
nx.draw(G, pos, node_color='skyblue', with_labels=True, node_size=3000, font_size=10)


# Show the plot
plt.title('Network Graph')
plt.show()
```
